```python
sentence=str(curpus[0])
sentence = sentence.lower()
cleanr = re.compile('[&:()+!!.,]')
cleantext = re.sub(cleanr, '', sentence)
rem_num = re.sub('[0-9]+', '', cleantext)
tokenizer = RegexpTokenizer(r'\w+')
tokens = tokenizer.tokenize(rem_num) 
filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
#filtered_words
```

```python
 def normalize_docuement(txt):
        
    txt=str(txt) # Coverting into string dtype
    
    txt = txt.lower() # Converting into lowercase
    
    # removing special_characters
    special_characters = re.compile('<.*?:>')
    cleanedtext = re.sub(special_characters, '', txt)
    
    
    # Removing Numbers
    removedNum = re.sub('[0-9]+', '', cleanedtext)
    
    # Removing stop words
    WordPunctTokenizer = RegexpTokenizer(r'\w+')
    tokens = WordPunctTokenizer.tokenize(removedNum)  
    filtered_words = [w for w in tokens if len(w) > 2 if not w in stopwords.words('english')]
    
    return " ".join(filtered_words)


Data['cleanText']=Data['Reviews'].map(lambda s:normalize_docuement(s)) 
```
